## FEATURE ENGINEERING

```python
df["sales"]=df["Quantity"]*df["UnitPrice"]
```

```python
sales_per_product=df.groupby("Description")["sales"].sum().head(10)
```

```python
sales_per_product
```

## DATA VISUALIZATION

```python
sales_per_product.plot(kind="bar")
plt.title("Top 10 product by sales")
plt.xlabel("product")
plt.ylabel("sales")
```

```python
plt.figure(figsize=(20,15))
sales_per_product.plot(kind="line")
plt.title("Top 10 product by sales")
plt.xlabel("product")
plt.ylabel("sales")

```
